```python
import streamlit as st
import numpy as np
```

```python
# Fonction pour optimiser le pipeline
def optimize_pipeline(df):
    df = df.drop_duplicates()
    df['CustomerEmail'].fillna('unknown@example.com', inplace=True)
    df['SupplierContact'].fillna('N/A', inplace=True)
    df['ProductPrice'] = pd.to_numeric(df['ProductPrice'], errors='coerce')
    df['AveragePricePerUnit'] = df['NetAmount'] / df['QuantitySold']
    return df

# Application principale
def main():
    st.title("Optimisation d'un Pipeline de Données")
    st.sidebar.header("Menu")
    
    options = st.sidebar.radio(
        "Options",
        ["Charger des données", "Analyser les données", "Optimiser le pipeline", "Visualiser les données"]
    )
    
    # Chargement des données
    if options == "Charger des données":
        st.subheader("Charger vos données")
        uploaded_file = st.file_uploader("Chargez un fichier CSV", type=["csv"])
        
        if uploaded_file is not None:
            df = pd.read_csv(uploaded_file)
            st.session_state['data'] = df
            st.success("Fichier chargé avec succès !")
    
    # Analyse des données
    if options == "Analyser les données":
        if 'data' in st.session_state:
            df = st.session_state['data']
            st.subheader("Analyse des données")
            st.write("Aperçu des données :")
            st.dataframe(df.head())
            st.write("Statistiques descriptives :")
            st.write(df.describe())
        else:
            st.warning("Veuillez d'abord charger des données.")
    
    # Optimisation du pipeline
    if options == "Optimiser le pipeline":
        if 'data' in st.session_state:
            df = st.session_state['data']
            optimized_df = optimize_pipeline(df)
            st.session_state['optimized_data'] = optimized_df
            st.success("Pipeline optimisé avec succès !")
        else:
            st.warning("Veuillez d'abord charger des données.")
    
    # Visualisation des données
    if options == "Visualiser les données":
        if 'optimized_data' in st.session_state:
            st.subheader("Données optimisées")
            optimized_df = st.session_state['optimized_data']
            st.dataframe(optimized_df.head())
            st.write("Histogramme d'une colonne numérique")
            numeric_columns = optimized_df.select_dtypes(include=[np.number]).columns
            column = st.selectbox("Choisissez une colonne :", numeric_columns)
            if column:
                st.bar_chart(optimized_df[column])
        else:
            st.warning("Veuillez optimiser les données d'abord.")

if __name__ == "__main__":
    main()
```
